```python
import numpy as np
import pandas as pd
```

```python
# first let's create a pandas Series
data = np.array(['a', 'b', 'c', 'd', 'e'])
series = pd.Series(data)
print("Pandas Series:")
print(series)
```

```python
# how to flatten a series like this :-

# Original Series of list
# 0    [Red, Green, White]
# 1           [Red, Black]
# 2               [Yellow]

ser = pd.Series([['Red', 'Green', 'White'], ['Red', 'Black'], ['Yellow']])

ser
# Flattening the Series
flat = [el for arr in ser for el in arr]
flat_series = pd.Series(flat)
print("\nFlattened Series:")
print(flat_series)
```
