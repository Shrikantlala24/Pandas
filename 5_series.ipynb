### 1. Series Creation

```python
import numpy as np
import pandas as pd

my_index = ['X','Y','Z','A','B']
my_data = [100,200,300,400,500]

# 1. creating series from lists
data = pd.Series(data=my_data, index=my_index)
print(data,'\n')

# 2. creating series from dictionary
my_dict = {'x':100, 'y':200,'z':300}
data_2 = pd.Series(my_dict)

print(data_2,'\n')

# 3. creating series from numpy array
arr = np.array([10,20,30,40,50])
data_3 = pd.Series(arr)

print(data_3)
```
