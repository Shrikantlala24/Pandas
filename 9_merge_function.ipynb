# Merge 
we use it to merge two dataframes based on one common ID, just like Join in SQL (inner, outer).
we'll find the orange cap players of each season (player with highest runs in each season)


```python
import numpy as np
import pandas as pd

matches = pd.read_csv(r'C:\Users\HP\Downloads\archive\matches.csv')
matches.head(3)
```

```python
delivery = pd.read_csv(r'C:\Users\HP\Downloads\archive\deliveries.csv')
delivery.head(3)
```

```python
common_df = delivery.merge(matches, left_on='match_id', right_on='id')
common_df.head(3)
```

```python
main_ser = common_df.groupby(['season', 'batsman']).batsman_runs.sum().sort_index()
main_ser.reset_index().sort_values('batsman_runs', ascending=False)
```
