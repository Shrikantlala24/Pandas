# Killer functions
here we'll learn some of the best methods/functions that we'll be using most of the times, whenever we are working on data

1. value_count()
2. sort_values()
3. drop_duplicates()
4. groupby()

```python
import numpy as np
import pandas as pd

data = pd.read_csv('Files/matches.csv')
data.head(2)

# click on the run button to execute the code cell and see the output
```


### 1. Value_counts()
we use it to count no. of frequencies of categories in categorical columns


```python
bhalue = data['winner'].value_counts()
bhalue
```

### 2. Sort_values()
here you sort values, you can also turn them around in ascending order by passing 'ascending = False', inside the function as argument

```python
print(bhalue.sort_values(ascending = True))

# let's apply the sort_values() function in dataframes
data.sort_values('date').head()
```

### 3. Drop_duplicates()
This function is used to take the first instance of each category and drop the rest of the repetitons of the category.

```python
data.sort_values(['season','date']).drop_duplicates('season')
```

### 4. Groupby()
This function is used to compile all the categories and grouping the same catgoryu elements.
