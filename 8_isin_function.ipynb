# isin() function :
here we have a dataset called deliveries, where you have each rows as the balls of any particular match, season, done by a batsman in which team. we have to find :-
1. best death over batsman in the history of IPL (death overs = 16-20)
2. best = high strike rate {strike rate = (no_runs / no_balls) / 100 }
3. batsman should have played atleast 200 balls

```python
import numpy as np
import pandas as pd

delivery = pd.read_csv(r'C:\Users\HP\Downloads\archive\deliveries.csv')
delivery.head(5)
```

```python
death_over_wale_samples = delivery[delivery.over > 15]

no_of_balls_har_player_ke = death_over_wale_samples.groupby('batsman').ball.count()
player_list = list(no_of_balls_har_player_ke[no_of_balls_har_player_ke > 200].index)
player_list
```

```python
final_dataframe = death_over_wale_samples[death_over_wale_samples['batsman'].isin(player_list)]
fodu_player_runs = final_dataframe.groupby('batsman').batsman_runs.sum()
fodu_player_balls = final_dataframe.groupby('batsman').ball.sum()
strike_rate = (fodu_player_runs/fodu_player_balls)
strike_rate.idxmax()
```
